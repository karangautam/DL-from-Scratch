# Gradient Descent with multiple inputs

```python
def w_sum(a,b):
    assert(len(a)==len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i]*b[i])
    return output
```

```python
def neural_network(input,weights):
    pred = w_sum(input,weights)
    return pred
```

```python
def ele_mul(a,b):
    c=[]
    for i in range(len(b)):
        c.append( a[i]*b[i])
    
    return c
```

```python
def scalar_ele_mul(a,b):
    c=[]
    for i in range(len(b)):
        c.append( a*b[i])
    
    return c
```

```python
weights = [0.1,0.2,-0.1]
toes = [8.5,9.5,9.9,9.0]
wlrec = [0.9,0.8,0.8,0.9]
nfans = [1.2,1.3,0.5,1]
result = [1,1,0,1]
input = [toes[0],wlrec[0],nfans[0]]
output = result[0]
pred = neural_network(input,weights)
error = (pred-output)**2
delta = (pred-output)

weight_delta = scalar_ele_mul(delta,input)
```

```python
for i in range(len(weight_delta)):
    assert(len(weights)==len(weight_delta))
    weights[i] -= weight_delta[i]
    print('weights {} weight_delta {}'.format(weights,weight_delta))
```

## putting it together with alpha

```python
weights = [0.1,0.2,-0.1]
toes = [8.5,9.5,9.9,9.0]
wlrec = [0.9,0.8,0.8,0.9]
nfans = [1.2,1.3,0.5,1]
result = [1,1,0,1]
input = [toes[0],wlrec[0],nfans[0]]
output = result[0]
iterations = 100
alpha = 0.01
for j in range(iterations):
    
    pred = neural_network(input,weights)
    error = (pred-output)**2
    delta = (pred-output)

    weight_delta = scalar_ele_mul(delta,input)
    for i in range(len(weight_delta)):
        assert(len(weights)==len(weight_delta))
        weights[i] -= weight_delta[i] * alpha
        #print('weights {} weight_delta {}'.format(weights,weight_delta))
    print('iteration {} prediction {} actual {} error {} weights {}'.format(j,pred,output,error,weights))
```
